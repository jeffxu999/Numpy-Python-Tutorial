```
# Zeroes & Ones - Numpy Python Tutorial

#######################################################

import numpy as np

giant_zeros_vector = np.zeros(100)

print(giant_zeros_vector)

giant_zeros_matrix = np.zeros((100,100))

print(giant_zeros_matrix)

#######################################################

giant_ones_vector = np.ones(100)

print(giant_ones_vector)

giant_ones_vector = np.ones((100,100))

print(giant_ones_vector)

#######################################################

giant_ones_matrix = np.ones((100,100))

print(giant_ones_matrix)

giant_15_matrix = giant_ones_matrix * 15

print(giant_15_matrix)
```
