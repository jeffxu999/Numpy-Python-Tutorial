```
# Random - Numpy Python Tutorial

#######################################################

import numpy as np
import matplotlib.pyplot as plt

random_rand = []
for i in range(1000):
    random_rand.append(np.random.rand())
    
print(random_rand)

plt.scatter(random_rand,range(len(random_rand)))

plt.xlabel("Random Between 0 - 1")

plt.show()

#######################################################

random_randint = []
for i in range(1000):
    random_randint.append(np.random.randint(1,5+1))
    
print(random_randint)

plt.scatter(random_randint,range(len(random_randint)))

plt.xlabel("Random Integers Dist")

plt.show()

#######################################################

random_normal = []
for i in range(1000):
    random_normal.append(np.random.normal(100,20))
    
print(random_normal)

plt.scatter(random_normal,range(len(random_normal)))

plt.xlabel("Random Normal Dist")

plt.show()
```
