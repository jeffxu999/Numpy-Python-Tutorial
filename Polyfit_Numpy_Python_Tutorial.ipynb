```
# Polyfit - Numpy Python Tutorial

#######################################################

import numpy as np
import matplotlib.pyplot as plt

x = [i for i in range(11)]

y = [i for i in range(0,21,2)]

print(x)
print(y)

slope, intercept = np.polyfit(x,y,1)

print(slope)

print(intercept.round(0))

plt.plot(x,y)

plt.show()

#######################################################

x = [i+np.random.randint(-10,11) for i in range(100)]

y = [i+np.random.randint(-10,11) for i in range(100)]

slope, intercept = np.polyfit(x,y,1)

print(slope)

print(intercept.round(0))

plt.scatter(x,y)

plt.show()

#######################################################

plt.scatter(x,y)

x_reg = [i for i in range(100)]
y_reg = [i*slope+intercept for i in range(100)]

plt.plot(x_reg,y_reg)

plt.show()
```
