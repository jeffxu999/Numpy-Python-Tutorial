```
# Meshgrid - Numpy Python Tutorial

#######################################################

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10)
y = np.arange(11,12)
x_mesh, y_mesh = np.meshgrid(x,y)

print(x)
print(y)

print(x_mesh)
print(y_mesh)

plt.scatter(x_mesh,y_mesh)

plt.show()

#######################################################

x = np.arange(1,10)
y = np.arange(11,13)
x_mesh, y_mesh = np.meshgrid(x,y)

print(x)
print(y)

print(x_mesh)
print(y_mesh)

plt.scatter(x_mesh,y_mesh)

plt.show()

#######################################################

x = np.arange(1,50)
y = np.arange(100,151)
x_mesh, y_mesh = np.meshgrid(x,y)

print(x)
print(y)

print(x_mesh)
print(y_mesh)

plt.scatter(x_mesh,y_mesh)

plt.show()
```
